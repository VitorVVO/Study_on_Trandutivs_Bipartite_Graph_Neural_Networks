## create dirs

```python
 # cria dirrotio para keyphrase
try:
    os.makedirs(output_path+"/mean/")
except OSError as e:
    if e.errno != errno.EEXIST:
        raise
```

```python
 # cria dirrotio para keyphrase
try:
    os.makedirs(output_path+"/std/")
except OSError as e:
    if e.errno != errno.EEXIST:
        raise
```

## Loads

```python
df_list = []
for i in range(10):
    aux_list = []
    aux_list.append(pd.read_pickle(input_path+str(i)+"/df_acc.pkl")) # acuracia
    aux_list.append(pd.read_pickle(input_path+str(i)+"/df_pre.pkl")) # precision
    aux_list.append(pd.read_pickle(input_path+str(i)+"/df_rec.pkl")) # recall
    aux_list.append(pd.read_pickle(input_path+str(i)+"/df_f1.pkl"))  # f1-score
    aux_list.append(pd.read_pickle(input_path+str(i)+"/df_sup.pkl")) # support
    aux_list.append(pd.read_pickle(input_path+str(i)+"/df_time.pkl")) # support
    df_list.append(aux_list)
```

```python
for i in range(10):
    for j in range(6):
        df_list[i][j].drop(['80%'], axis=1,inplace=True)
        df_list[i][j] = df_list[i][j].astype(float)
```

```python
df_list[0][0]
```

## Media

```python
# Calculate mean
mean_df_acc = pd.concat([df_list[0][0],df_list[1][0],df_list[2][0],df_list[3][0],df_list[4][0],df_list[5][0],df_list[6][0],df_list[7][0],df_list[8][0],df_list[9][0]]).groupby(level=0).mean()
mean_df_pre = pd.concat([df_list[0][1],df_list[1][1],df_list[2][1],df_list[3][1],df_list[4][1],df_list[5][1],df_list[6][1],df_list[7][1],df_list[8][1],df_list[9][1]]).groupby(level=0).mean()
mean_df_rec = pd.concat([df_list[0][2],df_list[1][2],df_list[2][2],df_list[3][2],df_list[4][2],df_list[5][2],df_list[6][2],df_list[7][2],df_list[8][2],df_list[9][2]]).groupby(level=0).mean()
mean_df_f1  = pd.concat([df_list[0][3],df_list[1][3],df_list[2][3],df_list[3][3],df_list[4][3],df_list[5][3],df_list[6][3],df_list[7][3],df_list[8][3],df_list[9][3]]).groupby(level=0).mean()
mean_df_sup = pd.concat([df_list[0][4],df_list[1][4],df_list[2][4],df_list[3][4],df_list[4][4],df_list[5][4],df_list[6][4],df_list[7][4],df_list[8][4],df_list[9][4]]).groupby(level=0).mean()
mean_df_time= pd.concat([df_list[0][5],df_list[1][5],df_list[2][5],df_list[3][5],df_list[4][5],df_list[5][5],df_list[6][5],df_list[7][5],df_list[8][5],df_list[9][5]]).groupby(level=0).mean()
```

```python
# Saving dataframes
mean_df_acc.to_csv(output_path+"/mean/df_acc.csv",index=True, index_label='Dataframe')   # acuracia
mean_df_pre.to_csv(output_path+"/mean/df_pre.csv",index=True, index_label='Dataframe')   # precision
mean_df_rec.to_csv(output_path+"/mean/df_rec.csv",index=True, index_label='Dataframe')   # recall
mean_df_f1.to_csv(output_path+"/mean/df_f1.csv",index=True, index_label='Dataframe')     # f1-score
mean_df_sup.to_csv(output_path+"/mean/df_sup.csv",index=True, index_label='Dataframe')   # support
mean_df_time.to_csv(output_path+"/mean/df_time.csv",index=True, index_label='Dataframe') # time
```

## STD

```python
# Calculate desvio padrao
std_df_acc = pd.concat([df_list[0][0],df_list[1][0],df_list[2][0],df_list[3][0],df_list[4][0],df_list[5][0],df_list[6][0],df_list[7][0],df_list[8][0],df_list[9][0]]).groupby(level=0).std()
std_df_pre = pd.concat([df_list[0][1],df_list[1][1],df_list[2][1],df_list[3][1],df_list[4][1],df_list[5][1],df_list[6][1],df_list[7][1],df_list[8][1],df_list[9][1]]).groupby(level=0).std()
std_df_rec = pd.concat([df_list[0][2],df_list[1][2],df_list[2][2],df_list[3][2],df_list[4][2],df_list[5][2],df_list[6][2],df_list[7][2],df_list[8][2],df_list[9][2]]).groupby(level=0).std()
std_df_f1  = pd.concat([df_list[0][3],df_list[1][3],df_list[2][3],df_list[3][3],df_list[4][3],df_list[5][3],df_list[6][3],df_list[7][3],df_list[8][3],df_list[9][3]]).groupby(level=0).std()
std_df_sup = pd.concat([df_list[0][4],df_list[1][4],df_list[2][4],df_list[3][4],df_list[4][4],df_list[5][4],df_list[6][4],df_list[7][4],df_list[8][4],df_list[9][4]]).groupby(level=0).std()
std_df_time= pd.concat([df_list[0][5],df_list[1][5],df_list[2][5],df_list[3][5],df_list[4][5],df_list[5][5],df_list[6][5],df_list[7][5],df_list[8][5],df_list[9][5]]).groupby(level=0).std()
```

```python
# Saving dataframes
std_df_acc.to_csv(output_path+"/std/df_acc.csv",index=True, index_label='Dataframe')   # acuracia
std_df_pre.to_csv(output_path+"/std/df_pre.csv",index=True, index_label='Dataframe')   # precision
std_df_rec.to_csv(output_path+"/std/df_rec.csv",index=True, index_label='Dataframe')   # recall
std_df_f1.to_csv(output_path+"/std/df_f1.csv",index=True, index_label='Dataframe')     # f1-score
std_df_sup.to_csv(output_path+"/std/df_sup.csv",index=True, index_label='Dataframe')   # support
std_df_time.to_csv(output_path+"/std/df_time.csv",index=True, index_label='Dataframe') # time
```
